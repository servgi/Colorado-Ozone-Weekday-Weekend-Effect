```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the relevant data

# Step 1: Extract columns related to Denver NO2 and day of week
DenverNOx = [col for col in df.columns if 'DENVER - CAMP_MDA1_NO2' in col]
DenverDay = [col for col in df.columns if 'dayofweek' in col]
Denver = DenverNOx + DenverDay

# Step 2: Select relevant columns from the DataFrame
posterdf = df[Denver]

# Step 3: Define a function to remove NaN values from arrays
def remove_nans(arr):
    return arr[~np.isnan(arr)]

# Apply the function to your DataFrame to remove NaNs
posterdf = posterdf.apply(remove_nans)

# Assuming you have custom mappings for dayofweek (e.g., 0 for Sunday, 3 for Wednesday)
wednesday_data = posterdf[posterdf['dayofweek'] == 3]  # Filtering Wednesday data
sunday_data = posterdf[posterdf['dayofweek'] == 0]     # Filtering Sunday data

# Step 4: Calculate averages for NO2 on Wednesdays and Sundays
wednesday_averages = wednesday_data.mean()
sunday_averages = sunday_data.mean()

# Step 5: Plotting the averages
fig, ax = plt.subplots()
ax.plot(wednesday_averages, marker="o", color='skyblue', label="Wednesday")
ax.plot(sunday_averages, marker="o", color='peachpuff', label="Sunday")

# Step 6: Adding plot labels and legend
ax.set_title("Average NO2 Levels on Wednesday and Sunday in Denver")
ax.set_xlabel('Columns')
ax.set_ylabel('Average Value')
ax.legend()

# Step 7: Showing the plot
plt.show()
```
